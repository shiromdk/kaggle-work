# Univariate Plotting with Pandas

In this section we will learn the basic pandas plotting facilities, starting with the simplest type of visualization: single-variable or "univariate" visualizations. This includes basic tools like bar plots and line charts

```python
## Show matlab plots inline
%matplotlib inline
import pandas as pd
reviews = pd.read_csv("winemag-data_first150k.csv", index_col=0)
reviews.head(3)
```

## Bar Charts and categorical data
Bar charts map categories to numbers

```python
reviews['province'].value_counts().head(10).plot.bar()
```

This bar chart tells us that California produces more wine than any another province. This bar chart tells us absolute numbers, but it's more useful to know relative proportions. 

```python
(reviews['province'].value_counts().head(10) / len(reviews)).plot.bar()
```

Bar charts are very flexible: The height can represent anything, as long as it is a number. And each bar can represent anything, as long as it is a category.

```python
reviews['points'].value_counts().sort_index().plot.bar()
```

In this case the categories are nominal categories: "pure" categories that don't make a lot of sense to order. Nominal categorical variables include things like countries, ZIP codes, etc.

## LINE CHARTS
A line chart can pass over any number of many individual values, making it the tool of first choice for distributions with many unique values or categories.

However, line charts have an important weakness: unlike bar charts, they're not appropriate for nominal categorical data. While bar charts distinguish between every "type" of point line charts mushes them together. So a line chart asserts an order to the values on the horizontal axis, and the order won’t make sense with some data.

```python
reviews['points'].value_counts().sort_index().plot.line()
```

## Area Charts
Area charts are line charts with the bottom filled in

```python
reviews['points'].value_counts().sort_index().plot.area()
```

## Interval Data
An interval variable goes beyond an ordinal categorical variable: it has a meaningful order, in the sense that we can quantify what the difference between two entries is itself an interval variable.
Line charts work well for interval data. Bar charts don't — unless your ability to measure it is very limited, interval data will naturally vary by quite a lot.

### Histogram

```python
reviews[reviews['price'] < 200]['price'].plot.hist()
```

a histogram is special kind of bar plot that splits your data into even intervals and displays how many rows are in each interval with bars. The only analytical difference is that instead of each bar representing a single value, it represents a range of values.

However, histograms have one major shortcoming. Because they break space up into even intervals, they don't deal very well with skewed data

```python
reviews['price'].plot.hist()
```
